4. Implicit fields initialization  คืออะไร

*คำตอบ*

การกำหนดค่าเริ่มต้นโดยอัตโนมัติ ให้กับฟิลด์ในคลาส เมื่อไม่ได้กำหนดค่าเริ่มต้นให้ฟิลด์นั้นๆ อย่างชัดเจนในโค้ด

5. Explicit field initialization คืออะไร



*คำตอบ*

การกำหนดค่าเริ่มต้นให้กับฟิลด์ในคลาสโดยตรงภายในโค้ด ซึ่งจะทำให้ฟิลด์เหล่านั้นมีค่าที่กำหนดไว้ตั้งแต่เริ่มต้น โดยไม่ต้องรอให้คอมไพเลอร์กำหนดค่าเริ่มต้นให้เอง

6. Implicit fields initialization  และ Explicit field initialization แตกต่างกันอย่างไร



*คำตอบ*

1. Implicit Field Initialization: ฟิลด์จะได้รับค่าเริ่มต้นโดยอัตโนมัติจาก C# ตามประเภทข้อมูล เช่น 0 สำหรับ int และ null สำหรับ string เมื่อไม่ได้กำหนดค่าเริ่มต้นในโค้ด
2. Implicit Field Initialization: ฟิลด์จะได้รับค่าเริ่มต้นโดยอัตโนมัติจาก C# ตามประเภทข้อมูล เช่น 0 สำหรับ int และ null สำหรับ string เมื่อไม่ได้กำหนดค่าเริ่มต้นในโค้ด
