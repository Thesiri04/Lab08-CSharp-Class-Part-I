4. Field และ Variable แตกต่างกันอย่างไร


*คำตอบ*

1. Field ประกาศภายในคลาส (แต่ภายนอกเมทอด) สามารถเข้าถึงจากเมทอดภายในคลาสหรือคลาสที่สืบทอด
2. Variable ประกาศภายในเมทอด ใช้ได้เฉพาะในบล็อกโค้ดที่ประกาศตัวแปรนั้น

5. เมื่อไหร่ควรใช้ Field และเมื่อไหร่ควรใช้ Variable

*คำตอบ*

6. ในโปรแกรมการทดลองนี้ มี field หรือไม่ อยู่ในบรรทัดที่เท่าไหร่


*คำตอบ*

1. ใช้ Field เมื่อค้องการเก็บข้อมูลที่เป็นส่วนหนึ่งของสถานะของ Object หรือคลาส และข้อมูลนั้นจะคงอยู่ตลอดชีวิตของ Object หรือคลาส
2. ใช้ Variable เมื่อต้องการเก็บข้อมูลชั่วคราวในช่วงเวลาสั้นๆ ที่มีการประมวลผลในเมทอดหรือบล็อกโค้ด

7. ผลจากการรันโปรแกรมใน Console Application และใน polyglot notebook เหมือนหรือแตกต่างกันอย่างไร
