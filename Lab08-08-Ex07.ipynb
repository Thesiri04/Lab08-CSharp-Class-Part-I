## คำถาม

1. ให้เขียน method ที่รับ parameter เป็น string และส่งกลับค่าความยาวของ string นั้น

````
using System;

class MyProgram
{
    public int add(int a, int b)
    {
        return a + b;
    }

    public int multiply(int a, int b)
    {
        return a * b;
    }

    // Method รับ string และคืนค่าความยาวของ string
    public int GetStringLength(string input)
    {
        return input.Length;
    }
}

var adder = new MyProgram();
int x = 2, y = 3;
int add_result = adder.add(x, y);
Console.WriteLine($"{x} + {y} = {add_result}");

int i = 2, j = 3;
int multiply_result = adder.multiply(i, j);
Console.WriteLine($"{i} x {j} = {multiply_result}");

string testString = "Hello";
int length = adder.GetStringLength(testString);
Console.WriteLine($"ความยาวของ \"{testString}\" คือ {length}");
````

*คำตอบ*

2. ให้สร้างโปรแกรมสูตรคูณ โดยประยุกต์ใช้เมธอด  `public int multiply(int a, int b)`


````
using System;

class MyProgram
{
    // เมธอดคูณ
    public int multiply(int a, int b)
    {
        return a * b;
    }
}

// สร้างอ็อบเจ็กต์ของ MyProgram
var adder = new MyProgram();

// เลือกตัวเลขที่ต้องการแสดงผลสูตรคูณ
int number = 7; // เช่น 7
for (int i = 1; i <= 12; i++)
{
    int result = adder.multiply(number, i);
    Console.WriteLine($"{number} x {i} = {result}");
}
````

3. จากโปรแกรมในคำถามข้อที่ 2 ให้สร้างเป็นสูตรคูณ แม่ 2 - 12

````
using System;

class MyProgram
{
    // เมธอดคูณ
    public int multiply(int a, int b)
    {
        return a * b;
    }
}

// สร้างอ็อบเจ็กต์ของ MyProgram
var adder = new MyProgram();

// แสดงสูตรคูณแม่ 2 ถึง 12
for (int number = 2; number <= 12; number++)
{
    Console.WriteLine($"สูตรคูณแม่ {number}:");
    for (int i = 1; i <= 12; i++)
    {
        int result = adder.multiply(number, i);
        Console.WriteLine($"{number} x {i} = {result}");
    }
    Console.WriteLine(); // เพิ่มบรรทัดเว้นระยะหลังจากแต่ละแม่
}
````
