## คำถาม

1. Access modifier ในภาษา C# คืออะไร


*คำตอบ*

คำสั่งที่ใช้ในการกำหนดสิทธิในการเข้าถึง

2. จงอธิบายขอบเขตและการใช้งานของ Access modifier ต่อไปนี้ 


2.1  public


*คำตอบ*

สมาชิกสามารถเข้าถึงได้จากทุกที่ภายนอกคลาส, เอาท์ไซด์โปรเจกต์


2.2 private


*คำตอบ*

สมาชิกสามารถเข้าถึงได้เฉพาะภายในคลาสหรือสตรัคเจอร์เดียวกัน

2.3 protected


*คำตอบ*

สมาชิกสามารถเข้าถึงได้ภายในคลาสเดียวกันและคลาสที่สืบทอดมาจากคลาสนี้

2.4 internal


*คำตอบ*

สมาชิกสามารถเข้าถึงได้ภายใน assembly เดียวกัน (โปรเจกต์เดียวกัน)

2.5 protected internal

*คำตอบ*

สมาชิกสามารถเข้าถึงได้จากภายในคลาสเดียวกัน, คลาสที่สืบทอดมาจากคลาสนี้, หรือจาก assembly เดียวกัน
